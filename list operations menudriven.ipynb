```python
list_element=[]
choice=0
while(1):
    print("1.create a new list\n2.Insert\n3.Remove\n4.Append\n5.Len\n6.Pop\n7.Clear\n8.Exit")
    choice=int(input("enter a choice:"))
    if choice==1:
        ele=int(input("enter a value to insert:"))
        list_element.append(ele)
        print("the new element is",list_element)
    elif choice==2:
        index=int(input("enter the index value:"))
        ele=int(input("enter a number to insert in list:"))
        list_element.insert(index,ele)
        print("inserted successfully",list_element)
    elif choice==3:
        ele=int(input("enter a number to remove from list:"))
        if ele in list_element:
            list_element.remove(ele)
            print("removed  successfully"list_element)
        else:
            print("list is empty")
    elif choice==4:
        ele=int(input("enter a number to append to the list:"))
        list_element.append(ele)
        print("appended  successfully",list_element)
    elif choice==5:
         print("the length of the list is:",len(list_element))
    elif choice==6:
        index=int(input("enter a index value:"))
        list_element.pop(index)
    elif choice==7:
        list_element.clear()
        print("list cleared",list_element)
    elif choice==8:
        break
    else:
        print("invalid choice")

```
