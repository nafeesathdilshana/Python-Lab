```python
#Write a program that makes use of a function to display sine, cosine, polynomial and exponential curves.
import numpy as np
import pylab
import matplotlib.pyplot as plt
def plot_curves():
    # Generate data points
    x = np.linspace(0,20,1000)
    # Sine curve
    y_sine = np.sin(x)
   
    # Cosine curve
    y_cosine = np.cos(x)
   
    plt.plot(x, y_sine, label='Sine')
    plt.plot(x, y_cosine, label='cosine')
    plt.legend(loc='upper right')
   
    y_polynomial = x**2 - 3*x + 1
    y_exponential=np.exp(x)
   
    plt.plot(x, y_polynomial, label='Ploynomial')
    plt.plot(x, y_exponential, label='Exponential')
    plt.legend(loc='upper right')
    pylab.ylim(-1.5, 5)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Sine,Cosine,Ploynomial,Exponential Graph")
plot_curves()
```
